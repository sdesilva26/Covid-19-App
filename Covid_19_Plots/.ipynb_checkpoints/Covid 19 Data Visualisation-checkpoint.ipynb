## Data Acquisition
Here I am going to scrape the covid-19 data from the NHS England website.

NHS England publish their daily figures for deaths at the following site: https://www.england.nhs.uk/statistics/statistical-work-areas/covid-19-daily-deaths/

#### Notes on the data
From the website

"This section contains information on deaths of patients who have died in hospitals in England and had tested positive for COVID-19 at time of death. All deaths are recorded against the date of death rather than the date the deaths were announced. Interpretation of the figures should take into account the fact that totals by date of death, particularly for most recent days, are likely to be updated in future releases. For example as deaths are confirmed as testing positive for COVID-19, as more post-mortem tests are processed and data from them are validated. Any changes are made clear in the daily files.

These figures will be updated at 2pm each day and include confirmed cases reported at 5pm the previous day. Confirmation of COVID-19 diagnosis, death notification and reporting in central figures can take up to several days and the hospitals providing the data are under significant operational pressure. This means that the totals reported at 5pm on each day may not include all deaths that occurred on that day or on recent prior days.

These figures do not include deaths outside hospital, such as those in care homes. This approach makes it possible to compile deaths data on a daily basis using up to date figures."

Essentially, every day they post a new updated xlsx file with the up to date figures they have. So the figures from the previous few days are likely to change as some people die and only are confirmed to of had covid-19 post-mortem. Therefore, the most recent data release should be used.

First I am going to convert the address in the additional info dataframe to a address

```python
# I added in here NHS because that seems to help the Google Maps API find the correct address later
def address_constructor(datapoint):
    address = ""
    address_array = datapoint[['Address_Line_1',
 'Address_Line_4',
 'Address_Line_5',
 'Postcode']]
    
    for index, i in enumerate(address_array):
        if (index == len(address_array)-1) and (type(i) == str):
            address += i + ", NHS"
        elif type(i) == str:
            address += i + ", "
        
    return address
    
```

```python
data_trusts_additional['address'] = data_trusts_additional.apply(lambda x: address_constructor(x),
                                                                axis=1)
data_trusts_additional.head()
```

Now convert the address into a longitude and a latitude

```python
import os
os.environ['GOOGLE_API_KEY']='AIzaSyB45-xIjLn0P8TrvAWX7Tn1riFhNQOPX_k'

def google_address_from_address(datapoint,key):
    from geopy.geocoders import GoogleV3
    from geopy.extra.rate_limiter import RateLimiter
    import os
    
    geolocator = GoogleV3(api_key=os.getenv('GOOGLE_API_KEY'), 
                      domain="maps.google.co.uk")
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=0.5)
    location = geocode(datapoint['address'])

    if (location.latitude != None):
        datapoint['latitude'] = location.latitude
    if (location.longitude != None):
        datapoint['longitude'] = location.longitude
    if (location.address != None):
        datapoint['address'] = location.address
        
    return datapoint
```

### Age and Region
For these two data sets I want to have 4 time series for each group in the data.
1. Deaths that day (already in the dataset)
2. Cumulative total of deaths
3. Relative change of deaths from previous day
4. Percentage relative change of deaths from previous day


```python
array = [data_age.columns.values,
         ['deaths today', 'cumulative total', 'change', 'relative change']]

index = pd.MultiIndex.from_product(array, names=['Age group', 'Death data'])

df_age_master = pd.DataFrame(index = data_age.index, columns=index)
df_age_master
```

Start off by populating the data I already have which is the deaths today for each date

```python
df_age_master.loc[:, ('80+', 'deaths today')] = data_age['80+']
```

```python
for age_g in data_age.columns:
    df_age_master.loc[:, (age_g, 'deaths today')] = data_age[age_g]
```

```python
df_age_master.head()
```

Now create data for the change in deaths compared to the previous day

```python
for age_g in data_age.columns:
    df_age_master.loc[:, (age_g, 'change')] = data_age[age_g].diff()
```

```python
df_age_master.head()
```

Create data for the relative change in deaths compared to the previous day as a percentage

```python
for age_g in data_age.columns:
    df_age_master.loc[:, (age_g, 'relative change')] = data_age[age_g].pct_change()*100
```

```python
df_age_master['80+'].head()
```

And finally we want a cumuulative total

```python
data_age['80+'].values.cumsum()

for age_g in data_age.columns:
    df_age_master.loc[:, (age_g, 'cumulative total')] = data_age[age_g].values.cumsum()
```

```python
df_age_master.head()
```

Now lets do the same for the region data set

```python
array = [data_region.columns.values,
         ['deaths today', 'cumulative total', 'change', 'relative change']]

index = pd.MultiIndex.from_product(array, names=['Age group', 'Death data'])

df_region_master = pd.DataFrame(index = data_age.index, columns=index)
df_region_master.head()
```

```python
for region in data_region.columns:
    df_region_master.loc[:, (region, 'deaths today')] = data_region[region]
    df_region_master.loc[:, (region, 'change')] = data_region[region].diff()
    df_region_master.loc[:, (region, 'relative change')] = data_region[region].pct_change()*100
    df_region_master.loc[:, (region, 'cumulative total')] = data_region[region].values.cumsum()
```

```python
df_region_master.head()
```

#### Bokeh Interactive Plot
Lets plot the above data but on an iteractive plot.

First plot the cumulative data

```python
from bokeh.plotting import ColumnDataSource
```

```python
groups = df_age_master.columns.get_level_values(0).unique().values
pd.DataFrame(df_age_master.T.loc[(groups, 'deaths today'),:].T.values, columns=groups, index=df_age_master.index).head()
```

```python
src = ColumnDataSource(pd.DataFrame(df_age_master.T.loc[(groups, 'deaths today'),:].T.values, columns=groups, 
                                    index=df_age_master.index))
src
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool
from bokeh.palettes import Dark2_5 as palette
import itertools


# output to static HTML file
output_file("html_plots/age_group_cumulative.html")


# create a new plot with a title and axis labels
p1 = figure(title="Total cumulative deaths from Covid-19 in UK grouped by age", x_axis_label='Date',
           y_axis_label='Total deaths', x_axis_type="datetime")


src = ColumnDataSource(pd.DataFrame(df_age_master.T.loc[(groups, 'cumulative total'),:].T.values, columns=groups, 
                                    index=df_age_master.index))
colors = itertools.cycle(palette)

for group, color in zip(groups, colors):
    p1.line(x='index', y=group, source=src, line_width=1, legend_label=group, color=color)
    p1.circle(x='index', y=group, source=src, size=6, legend_label=group, color=color)

    hover = HoverTool(tooltips =[
     ('Date','@index{%F}'),('Total deaths','$y{0}')],
                     formatters={'@index': 'datetime'})

    p1.add_tools(hover)

p1.legend.location = "top_left"
p1.legend.click_policy="hide"

show(p1)
```
